```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import sklearn
from sklearn import linear_model
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.pyplot import rcParams
import pickle
from sklearn import sklearns.externals
rcParams['figure.figsize']=15,15
sb.set_style('whitegrid')
```

```python
df=pd.read_csv('cars.csv')
df
```

```python
new_df=df.drop(['Car','Origin'],axis=1)
new_df
```

```python
plt.xlabel('Displacement',fontsize=30)
plt.ylabel('Acceleration',fontsize=30)
plt.plot(df.Displacement,df.Acceleration,color='red',marker='+')
plt.show()
```

```python
plt.xlabel('Displacement',fontsize=30)
plt.ylabel('Acceleration',fontsize=30)
plt.scatter(df.Displacement,df.Acceleration,color='red',marker='+',lw='8')
plt.show()
```

```python
reg=linear_model.LinearRegression()
reg.fit(new_df[['MPG','Cylinders','Displacement','Horsepower','Weight']],new_df.Acceleration)
```

```python
print("coefficients for this equation will be ",reg.coef_)
```

```python
print("intercept for this equation will be ",reg.intercept_)
```

```python
print("Acceleration of given data will be",reg.predict([[18,8,307,130,3504]]))
```

```python
import pickle
with open('car_model','wb') as f:
    pickle.dump(reg,f)
with open('car_model','rb') as f:
    obj_pickle=pickle.load(f)
obj_pickle.predict([[18,8,307,130,3504]])
```

```python
from sklearn.externals import joblib
joblib.dump(reg,'car_model_joblib')
obj_joblib=joblib.load('car_model_joblib')
obj_joblib.predict([[18,8,307,130,3504]])
```

```python
test_df=new_df.drop(['Acceleration','Model'],axis=1)
result=pd.DataFrame(reg.predict(test_df))
result.columns=['Predicted_Acceleration']
final_df=pd.concat([df,result],axis=1)
final_df
```

```python
plt.xlabel('Displacement',fontsize=30)
plt.ylabel('Predicted_Acceleration',fontsize=30)
plt.plot(final_df.Displacement,final_df.Predicted_Acceleration)
plt.show()
```

```python
plt.xlabel('Displacement',fontsize=30)
plt.ylabel('Predicted_Acceleration',fontsize=30)
plt.scatter(final_df.Displacement,final_df.Predicted_Acceleration)
plt.show()
```
